<b>c) Add a new column with the name "price_category". Then, find out the quantile boundaries of house prices. If you have the quantile limits, you can give four different price categories to the houses. Houses with a price in the bottom 25% quantile receive category 1. Houses with a price between 25% and 50% quantile receive category 2. Houses with a price between 50% and 75% receive category 3 and houses with a price in the highest quantile receive category 4. Create a figure with four boxplots, where each boxplot refers to a price category (1-4). The y-axis of the boxplots should represent the distribution of living space, i.e. column "sqft_living".

```python
# YOUR CODE TO ANSWER THIS QUESTION (YOU CAN USE MULTIPLE CELLS)
bin_labels_5 = ['1', '2', '3', '4']
df['price_category'] = pd.qcut(df['price'],
                              q=4,
                              labels=bin_labels_5)
df
```

```python
sns.boxplot(x="price_category", y="sqft_living", data=df)
```
